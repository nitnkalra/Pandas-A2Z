## How to Create a Series?

### Series from list

````python
#string

country = ['india','pakistan','nepal','china']
pd.Series(country)
````

Every series object has 2 things: index (01,2,3,4) & value (india,pakistan,nepal.. etc)

dtype: object (pandas denote string as object dtype)

````python
#integers

runs = [1,2,3,4,5,6]
pd.Series(runs)
````

dtype: int64 for integer values

````python
#custom index
#instead of index auto creation by pandas like above, we can also pass custom index's

marks = [10,20,30,40,50]
sub = ['hindi','english','maths','science','history']

#marks as index & sub as values
pd.Series(marks,index = sub)
````

````python
#series name
score = pd.Series(marks,index = sub,name = 'Score')
score
````

### Series from dict

````python
marks = {
    'hindi':10,
    'english':20,
    'maths':30
}

marks_ser = pd.Series(marks,name = 'Score')
marks_ser
````

###`Q-5.`Write a function to change the data type of given a column or a Series. Function takes series and data type as input, returns the converted series.
```
series = pd.Series([1,2,'Python', 2.0, True, 100])
change to float type data
```
Note: Read about pd.to_numeric()

````python
series = pd.Series([1,2,'Python', 2.0, True, 100])

def func(ser,dtype):
  ser_new = pd.to_numeric(ser,errors='coerce')
  return ser_new.astype(dtype)

func(series,float)
````
