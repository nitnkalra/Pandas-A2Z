```python
#problem in vectorized operation in vanilla python

s = ['cat','mat',None,'rat']

# [i.startswith('c') for i in s]

#python doesnt work on None type data or missing data
#it also relatively slow then pandas
```

## Pandas String operations

```python
df = pd.read_csv('/content/titanic.csv')
df.head()
```

```python
df['Name']
```

```python
#lower/upper/title/capitalize

df['Name'].str.upper()
```

```python
df['Name'].str.title()
```

```python
#len/strip

df['Name'].str.len()
```

```python
df[df['Name'].str.len() == df['Name'].str.len().max()]
```

```python
df['Name'].str.strip()
```

```python
#split --> get

df['Name'].str. split(',').str.get(0)
```

```python
df['Name'].str. split(',').str.get(1)
```

```python
df['last_name'] = df['Name'].str. split(',').str.get(0)
df
```

```python
df['Name'].str. split(',').str.get(1).str.strip().str.split(' ')
```

```python
df[['title','fname']] = df['Name'].str. split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)
```

```python
df
```

```python
df['title'].value_counts()
```

```python
#replace

df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
```

```python
df['title'].value_counts()
```

```python
#filtering
#startswith/endswith

#name strats with A
df[df['Name'].str.startswith('A')]
```

```python
df[df['Name'].str.endswith('A')]
```

```python
#advance level filtering
#regex

# filter name contains 'john'
df[df['Name'].str.contains('John', case=True)]
```

```python
#name starts & ends with vowels

df.columns
```

```python
df[df['last_name'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```

```python
#slicing

df['Name'].str[::2]
df['Name'].str[::-1]
```
